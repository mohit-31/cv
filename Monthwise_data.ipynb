```
import pandas as pd
import numpy as np

df = pd.read_csv('tempal2.txt', sep=" ",header=None)
df.columns=('jan','f','mar',"ap","ma","jun","jul","aug","sep","oct","nov","dec","la")
df_min=pd.read_csv('min_temp.txt', sep=" ",header=None)
df_min.columns=('jan','f','mar',"ap","ma","jun","jul","aug","sep","oct","nov","dec","la")
```

```

df_w=pd.read_csv('wind_sp.txt', sep=" ",header=None)
df_w.columns=('jan','f','mar',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
df_w;
```

```

hum=pd.read_csv('humid.txt', sep=" ",header=None)
hum.columns=('jan','f','mar',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
hum;
```

```
cloud=pd.read_csv('cloud.txt', sep=" ",header=None)
cloud.columns=('jan','f','mar',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
cloud;
```

```
rain_mi=pd.read_csv('rain_min.txt', sep=" ",header=None)
rain_mi.columns=('jan','f','mar',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
rain_mi;
```

```
rainy_day=pd.read_csv('rainy_day.txt', sep=" ",header=None)
rainy_day.columns=('jan','f','mar',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
rainy_day;
```

```
df=df[['jan','f','ma',"ap","mar","jun","jul","aug","sep","oct","nov","dec"]]
df;
```

```
df_min=df_min[['jan','f','mar',"ap","ma","jun","jul","aug","sep","oct","nov","dec"]]
df_min;
```

```
column_names = ["max_temp(c)", "min_temp(c)", "mean_windspeed(km/hr)","re_hum(%)","cloud(okta)",
                "Mean_rain(mm)","avg._no._rainyday","district"]
z = pd.DataFrame(columns = column_names)
z["max_temp(c)"]=df['f']
z["min_temp(c)"]=df_min['f']
z["mean_windspeed(km/hr)"]=df_w['f']
z["re_hum(%)"]=hum['f']
z["cloud(okta)"]=cloud["f"]
z["district"]=["Bhagalpur","chapra","Darbhanga","Dehri","Forbesganj","Gaya","jamui","Motihari","Muzaffarpur","patna","Purnea","Raxaul","Sabour","Supaul"]
z["Mean_rain(mm)"]=rain_mi['f'];
z["avg._no._rainyday"]=rainy_day["f"]
z
```
